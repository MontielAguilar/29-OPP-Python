## 1) Crear una clase básica con atributos y métodos básicos
Escribe un programa en Python para crear una clase Vehicle con atributos de instancia __max_speed__ y __mileage__. 

Y 2 métodos:
<ul>
    <li>get_max_speed()</li>
    <li>increase_mileage(ammount_to_increase)</li>
</ul>

```python
class Vehicle:
    def __init__(self, max_speed, mileage):
        self.max_speed = max_speed
        self.mileage = mileage
    
    def get_max_speed(self):
        return self.max_speed
    
    def increase_mileage(self, amount_to_increase):
        self.mileage += amount_to_increase

# Crear un objeto de la clase Vehicle
car = Vehicle(180, 15000)

# Obtener la velocidad máxima del vehículo
print("Velocidad máxima del vehículo:", car.get_max_speed())  # Output: Velocidad máxima del vehículo: 180

# Incrementar el kilometraje del vehículo en 1000
car.increase_mileage(1000)
print("Kilometraje actual del vehículo:", car.mileage)  # Output: Kilometraje actual del vehículo: 16000
```

## 2) Crear una clase con atributos y métodos
Escribe una clase __Rectangle__ en lenguaje Python, que te permita construir un rectángulo con atributos de longitud y anchura.

Crea un método __perimeter()__ para calcular el perímetro del rectángulo y un método __area()__ para calcular el área del rectángulo.

Crear un método __display()__ que muestre la longitud, la anchura, el perímetro y el área de un objeto creado mediante una instanciación de la clase rectángulo.

Sobrescribir el método __\_str\_()__ para que la representación de la cadena por defecto de la clase sea la definida en display()

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def perimeter(self):
        return 2 * (self.length + self.width)
    
    def area(self):
        return self.length * self.width
    
    def display(self):
        print("Longitud:", self.length)
        print("Anchura:", self.width)
        print("Perímetro:", self.perimeter())
        print("Área:", self.area())
    
    def __str__(self):
        return f"Rectángulo de longitud {self.length} y anchura {self.width}. Perímetro: {self.perimeter()}, Área: {self.area()}"

# Crear un objeto de la clase Rectangle
rect = Rectangle(5, 10)

# Usar el método display para mostrar los detalles del rectángulo
rect.display()
# Output:
# Longitud: 5
# Anchura: 10
# Perímetro: 30
# Área: 50

# Utilizar el método __str__ para mostrar la representación de cadena del objeto Rectangle
print(str(rect))
# Output: Rectángulo de longitud 5 y anchura 10. Perímetro: 30, Área: 50
```

## 3) Crear una clase más avanzada con atributos y métodos
Crea una clase de Python llamada __BankAccount__ que represente una cuenta bancaria, teniendo como atributos 

<ul>
<li> <b>account_number</b> (de tipo numérico), <b>name</b> (nombre del titular de la cuenta, string), <b>balance</b>.</li>
</ul>

Y con los siguientes métodos:

<ul>
<li>Crear un constructor con parámetros: <b>account_number</b>, <b>name</b>, <b>balance</b>.</li>

<li>Crear un método <b>deposit(ammount)</b> que gestione las acciones de depósito.</li>

<li>Crear un método <b>withdrawal(ammount)</b> que gestione las acciones de retirada.</li>
    
<li>Crear un método <b>transfer(acc_to_transfer_to_obj, ammount)</b> que transfiera dinero a otro objeto de cuenta bancaria que reciba como parámetro.</li>

<li>Crear un método <b>apply_bank_fees(percentage)</b> que aplique las comisiones bancarias con un porcentaje del saldo actual de la cuenta.</li>
 
<li>Crear un método <b>display()</b> para mostrar los detalles de la cuenta.</li>
</ul>

Implementar también los métodos necesarios para comparar diferentes cuentas bancarias por saldo utilizando los operadores > < >= <= y ==.

```python
class BankAccount:
    def __init__(self, account_number, name, balance):
        self.account_number = account_number
        self.name = name
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        print(f"Depósito de {amount} realizado. Nuevo saldo: {self.balance}")
    
    def withdrawal(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Retirada de {amount} realizada. Nuevo saldo: {self.balance}")
        else:
            print("Saldo insuficiente para realizar la retirada.")
    
    def transfer(self, acc_to_transfer_to_obj, amount):
        if amount <= self.balance:
            self.balance -= amount
            acc_to_transfer_to_obj.deposit(amount)
            print(f"Transferencia de {amount} realizada a la cuenta {acc_to_transfer_to_obj.account_number}")
        else:
            print("Saldo insuficiente para realizar la transferencia.")
    
    def apply_bank_fees(self, percentage):
        fees = (percentage / 100) * self.balance
        self.balance -= fees
        print(f"Comisiones bancarias aplicadas. Nuevo saldo: {self.balance}")
    
    def display(self):
        print("Detalles de la cuenta:")
        print("Número de cuenta:", self.account_number)
        print("Nombre del titular:", self.name)
        print("Saldo:", self.balance)
    
    # Métodos de comparación
    def __lt__(self, other):
        return self.balance < other.balance
    
    def __le__(self, other):
        return self.balance <= other.balance
    
    def __eq__(self, other):
        return self.balance == other.balance
    
    def __ne__(self, other):
        return self.balance != other.balance
    
    def __gt__(self, other):
        return self.balance > other.balance
    
    def __ge__(self, other):
        return self.balance >= other.balance

# Ejemplo de uso
account1 = BankAccount(123456, "John Doe", 1000)
account2 = BankAccount(654321, "Jane Doe", 2000)

account1.deposit(500)
account1.withdrawal(200)
account1.transfer(account2, 300)
account1.apply_bank_fees(2)
account1.display()

if account1 > account2:
    print("La cuenta 1 tiene un saldo mayor que la cuenta 2.")
elif account1 < account2:
    print("La cuenta 1 tiene un saldo menor que la cuenta 2.")
else:
    print("Las cuentas tienen el mismo saldo.")
```

```python
class Vehicle:
    def __init__(self, name, mileage, capacity):
        self.name = name
        self.mileage = mileage
        self.capacity = capacity

    def fare(self):
        return self.capacity * 100

class Bus(Vehicle):
    def fare(self):
        # Tarifa base según la capacidad de asientos
        base_fare = super().fare()
        # Agregar el 10% extra para el mantenimiento de autobuses
        bus_maintenance_charge = 0.1 * base_fare
        total_fare = base_fare + bus_maintenance_charge
        return total_fare

# Crear un objeto de la clase Bus
School_bus = Bus("School Volvo", 12, 50)
print("Total Bus fare is:", School_bus.fare())  # Output: Total Bus fare is: 5500
```

```python
class FormulaError(Exception):
    pass

def calculate(expression):
    try:
        elements = expression.split()
        if len(elements) != 3:
            raise FormulaError("La fórmula debe constar de tres elementos: número, operador, número.")
        
        num1 = float(elements[0])
        operator = elements[1]
        num2 = float(elements[2])
        
        if operator not in ['+', '-']:
            raise FormulaError("El operador debe ser '+' o '-'.")
        
        if operator == '+':
            result = num1 + num2
        else:
            result = num1 - num2
        
        print(result)
        
    except ValueError:
        raise FormulaError("Los números deben ser valores numéricos válidos.")
    except FormulaError as e:
        print("Error:", e)

# Calculadora interactiva
while True:
    user_input = input("Ingrese una fórmula (por ejemplo, '1 + 1'): ")
    if user_input.lower() == 'quit':
        break
    calculate(user_input)
```

## 6) Manejo de excepciones
Continúa con el ejercicio anterior y crea ahora tres clases hijas de FormulaError:

<ul>
<li><b>FormulaErrorNumber</b>: La levantarás cuando el usuario introduzca un valor erróneo para los números</li>
<li><b>FormulaErrorOp</b>: Se lanzará cuando el usuario introduzca un valor erróneo para el operador</li>
<li><b>FormulaErrorArgs</b>: Se lanzará cuando el usuario introduzca un número incorrecto de argumentos a la calculadora</li>
</ul>

Levántalos en los casos descritos y manéjalos con try, excepto para que puedas mostrar mensajes personalizados al usuario dependiendo de lo que esté mal en la entrada.

Puntos extra: Si también implementas la operación de división intenta manejar el caso ZeroDivisionError también

```python
import re #https://www.youtube.com/watch?v=7QUmK6cW_Rg

# Definición de la clase base de excepción para errores en fórmulas
class FormulaError(Exception):
    pass

# Definición de las clases hijas de FormulaError para diferentes tipos de errores
class FormulaErrorNumber(FormulaError):
    pass

class FormulaErrorOp(FormulaError):
    pass

class FormulaErrorArgs(FormulaError):
    pass

# Función para calcular el resultado de una fórmula dada
def calcular(formula):
    try:
        # Utilizar expresiones regulares para dividir la entrada en número, operador, y número
        match = re.match(r"^\s*(-?(?:\d+(?:\.\d+)?|True|False))\s*([+\-*/])\s*(-?(?:\d+(?:\.\d+)?|True|False))\s*$", formula)
        if not match:
            raise FormulaErrorArgs("La fórmula debe constar de un número, un operador (+, -, *, /), y otro número separados por espacios.")

        # Extraer los elementos individuales de la fórmula
        num1 = match.group(1)
        operador = match.group(2)
        num2 = match.group(3)

        # Convertir los números a tipo float o booleano según corresponda
        if num1.lower() == 'true':
            num1 = True
        elif num1.lower() == 'false':
            num1 = False
        else:
            num1 = float(num1)

        if num2.lower() == 'true':
            num2 = True
        elif num2.lower() == 'false':
            num2 = False
        else:
            num2 = float(num2)

        # Realizar el cálculo según el operador
        if operador == '+':
            resultado = num1 + num2
        elif operador == '-':
            resultado = num1 - num2
        elif operador == '*':
            resultado = num1 * num2
        elif operador == '/':
            if num2 == 0:
                raise ZeroDivisionError("No se puede dividir por cero.")
            resultado = num1 / num2
        else:
            raise FormulaErrorOp("Operador inválido.")

        # Imprimir el resultado
        print(resultado)

    # Manejo de excepciones
    except (ValueError, TypeError):
        raise FormulaErrorNumber("Los números deben ser válidos.")
    except ZeroDivisionError as e:
        print(f"Error: {e}")
    except FormulaError as e:
        print(f"Error: {e}")

# Bucle principal para solicitar la entrada al usuario y calcular el resultado
while True:
    entrada = input("Introduce una fórmula (número operador número) o escribe 'quit' para salir: ")
    if entrada.lower() == 'quit':
        break
    else:
        try:
            calcular(entrada)
        except FormulaError as e:
            print(f"Error: {e}")
```
